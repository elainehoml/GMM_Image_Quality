# Grey value standardisation with Gaussian Mixture Models

Some image processing tasks affect the grey value distribution of image datasets, which can complicate comparison between images. Here, we present a method of using Gaussian Mixture Models (GMMs) to standardise grey value distributions of processed images.

```python
# Import libraries
import os
import sys
from IPython.display import display, display_html
import ipywidgets as widgets
import matplotlib.pyplot as plt
import seaborn as sns
from skimage import io
import numpy as np
import pandas as pd
sns.set()

# Import functions
root_dir = os.path.split(os.path.split(os.getcwd())[0])[0] # go two levels up
sys.path.append(os.path.join(root_dir, "main"))
import QM_fit
import QM_calc
```

## Test images

A test image of an empty wax block scanned with X-ray micro-computed tomography ($\mu$CT) was used. Common image processing tasks were performed on a cropped version of this image using Fiji/ImageJ.

* Conversion to 8-bit
* Fourier bandpass filtering
* Gaussian blurring $\sigma = 2.0$
* Rescaling to 0.5x in x, y and z-directions.


```python
# Load test images
filenames = os.walk(os.path.join(os.getcwd(), 'test_images'))
test_images = {}
for root, dirs, files in filenames:
    for file in files:
        if os.path.splitext(str(file))[-1] == '.tif':
            test_images[file] = io.imread(os.path.join(root, file))
print("{} test images loaded".format(len(test_images)))
```

```python
plt.figure(figsize = [8.3, 11.7])
for (n, file) in enumerate(test_images, start=1):
    plt.subplot(len(test_images), 2, 2*n - 1)
    plt.imshow(test_images[file][0], cmap='gray') # first slice only
    plt.xticks([])
    plt.yticks([])
    plt.title(file)
    
    plt.subplot(len(test_images), 2, 2*n)
    plt.hist(test_images[file].flatten(), bins=256, density=True)
    plt.xlim([0,300])
    plt.ylim([0,0.1])
    plt.xlabel('Grey value')
    plt.ylabel('Prob density')
    plt.title('')

plt.tight_layout()
```

## Fitting GMMs

Gaussian mixture models were fitted to these images, with three components each. 10% of each stack was used.

```python
# Fit GMMs

mu_wax = {}
mu_cassette = {}
sigma_wax = {}
sigma_cassette = {}

for image in test_images:
    print(image)
    mu, sigma = QM_fit.QM_fit(test_images[image], 2, img_fname="NA", save_results=False)
    mu_wax[image] = mu[0]
    mu_cassette[image] = mu[1]
    sigma_wax[image] = sigma[0]
    sigma_cassette[image] = sigma[0]
```

```python
# show table of fitted Gaussian properties


df = pd.DataFrame({'mu_wax':pd.Series(mu_wax),
                   'mu_cassette':pd.Series(mu_cassette),
                   'sigma_wax':pd.Series(sigma_wax),
                   'sigma_cassette':pd.Series(sigma_cassette)})
display(df)
```
